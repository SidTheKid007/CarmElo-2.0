```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt 
df = pd.read_csv('NBASeasonData.csv')
df.head()
```

```python
df.loc[df['Player'].str[-1:] == '*', 'Player'] = df['Player'].str[:-1]
df.loc[df['Pos'] != 'C', 'Pos'] = df['Pos'].str[:2]
df = df.replace('C-', 'C')
df.head()
```

```python
players = df.Player.unique()
```

```python
df.tail()
```

```python
cols = df.columns.values.tolist() + ['PTS2', 'TRB2', 'AST2', 'STL2', 'BLK2']
```

```python
ovr = []
for player in players:
    df1 = df[df['Player'] == player]
    df1 = df1.drop_duplicates(subset ='Year') 
    df1['PTS2'] = df1['PTS'].shift(-1)
    df1['TRB2'] = df1['TRB'].shift(-1)
    df1['AST2'] = df1['AST'].shift(-1)
    df1['STL2'] = df1['STL'].shift(-1)
    df1['BLK2'] = df1['BLK'].shift(-1)
    df1 = df1.dropna()
    playerdata = df1.values.tolist()
    ovr = ovr + playerdata
```

```python
import numpy as np
ovr2 = np.array(ovr)
refined = np.reshape(ovr, (-1, 29))
```

```python
df2 = pd.DataFrame(refined, columns=cols)
df2.head()
```

```python
len(df2[df2['Tm'] == 'TOT'])/len(df2)
```

```python
df2.to_csv('NBAFutures1.csv')
```
